```python
import torch
import torchvision
import torchvision.transforms as T
import torch.nn.functional as F
import numpy as np
import matplotlib.pyplot as plt
import cv2 
from PIL import Image
from torchvision import datasets, transforms
from torch.autograd import Variable
import torch.nn as nn
```

```python
class edgeDetector():
    def __init__(self, path):
        self.path = path
    
    def prepareImageBlur(self):
        image = Image.open(self.path)
        
        plt.imshow(image)
        plt.pause(0.001) # Needs to have this or next images don't show
        
        # The gaussian blur itself
        transform = T.GaussianBlur(kernel_size=(7), sigma=(5,8))
        img = transform(image)
        image_tensor = transforms.ToTensor()(img)
        image = image_tensor.expand(1,-1,-1,-1)
        self.image = image
        
        # Needs to be done to be able to show with imshow() because dimensions
        unloader = transforms.ToPILImage()
        blurred = image.squeeze(0)
        blurred = unloader(blurred)
        plt.imshow(blurred)
        plt.pause(0.001)
        
    def gradientsCalculations(self):
        img = self.image
        img = img.squeeze(0)
        tensor=torch.unbind(img)
        the_tensor=tensor[0].expand(1, 1,-1,-1)

        #Sobel x
        sobel_x=np.array([[1, 0, -1],[2,0,-2],[1,0,-1]])
        conv1=nn.Conv2d(1, 1, kernel_size=3, stride=1, padding=1, bias=False)
        conv1.weight=nn.Parameter(torch.from_numpy(sobel_x).float().expand(1, 1,-1,-1))
        gradient_x=conv1(Variable(the_tensor)).data.view(1,the_tensor.shape[2],the_tensor.shape[3])

        #Sobel y
        sobel_y=np.array([[1, 2, 1],[0,0,0],[-1,-2,-1]])
        conv2=nn.Conv2d(1, 1, kernel_size=3, stride=1, padding=1, bias=False)
        conv2.weight=nn.Parameter(torch.from_numpy(sobel_y).float().expand(1, 1,-1,-1))
        gradient_y=conv2(Variable(the_tensor)).data.view(1,the_tensor.shape[2],the_tensor.shape[3])

        gradients = torch.sqrt(gradient_x **2 + gradient_y **2)
        self.gradients = gradients
        
        # Needs to be done to be able to show with imshow() because dimensions
        unloader = transforms.ToPILImage()
        image = gradients.squeeze(0)
        image = unloader(image)
        
        plt.imshow(image)
        plt.pause(0.001) # Needs to have this or next images don't show
        
    def threshold_process(self):
        # Turn them gradients to grayscale
        img_gray = transforms.Grayscale()(self.gradients)
        
        # The value deciding what gets turned white and what gets turned black
        threshold_value = 0.1
        binary_mask = torch.where(img_gray >= threshold_value, torch.tensor(0.5), torch.tensor(0.9))
        plt.imshow(binary_mask.squeeze().numpy(), cmap='binary')

        plt.show()
        
```

```python
firstDetection = edgeDetector("pear.jpg")
firstDetection.prepareImageBlur()
firstDetection.gradientsCalculations()
firstDetection.threshold_process()
```
