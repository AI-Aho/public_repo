```python
from pandas import DataFrame 
from sklearn import linear_model
import matplotlib.pyplot as visualisation
import numpy as np
from sklearn.preprocessing import PolynomialFeatures
data = {'number_of_observations': [1,4,10,20,35,56,84],
        'month_n': [1,2,3,4,5,6,7]}
data_frame = DataFrame(data) 

polynomiaal_features = PolynomialFeatures(degree=2)
input_dimension_polynomic = polynomiaal_features.fit_transform(data_frame[['month_n']])
regression_tool = linear_model.LinearRegression()
regression_tool.fit(input_dimension_polynomic, data_frame['number_of_observations'])

visualisation.title('number_of_observations Vs month_n', fontsize=14)
visualisation.xlabel('month_n', fontsize=14)
visualisation.ylabel('number_of_observations', fontsize=14)
visualisation.grid(True) 

visualisation.scatter(data_frame['month_n'], data_frame['number_of_observations'])

regression_curve = regression_tool.predict(input_dimension_polynomic)

visualisation.plot(data_frame['month_n'], regression_curve, color='y')

visualisation.show()
from sklearn.metrics import r2_score
animals_r2_score=r2_score( data_frame['number_of_observations'],regression_curve)
##
print("R2 score:",animals_r2_score) 
```
