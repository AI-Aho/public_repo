### Dataframe 2


```python
name_age = pd.DataFrame({'Names':['Johan Kalac', 'Marina Tribble', 'Markus Ferdin', 'Julia Maria'], 'Age':[45,67,18,33], 'Year':[2020,2020,2020,2020]})
name_age.head()
```

```python
name_age['current_age'] = name_age.apply(lambda x: x['Age'] +2, axis=1)
name_age
```

```python
name_age['Names'] = name_age['Names'].str.upper()
name_age
```

```python
name_age["current_year"] = name_age.apply(lambda x: x['Year'] +2, axis=1)
name_age
```

```python
name_age["future_year"] = name_age.apply(lambda x: x['current_year'] +10, axis=1)
name_age
```

```python
name_age["future_age"] = name_age.apply(lambda x: x['current_age'] +10, axis=1)
name_age
```

```python
name_age['age_classification'] = np.where(name_age['current_age'] > 65, 'old', 'young')
name_age
```

### Dataframe 3

```python
import pandas as pd
import numpy as np
pacients = pd.DataFrame({'Name':['Han Solo', 'Indiana Jonnes', 'John McClane', 'James Bond', 'Don Vito Corleone'],'Age':[76,87,67,75,80],'Upper_pressure':[90,120,135,145,90],'Lower_pressure':[80,70,85,88,90],'Weight':[150,100,90,76,84]})
pacients.head()
```

```python
pacients['First_name'] = pacients['Name'].str.split().str[0]
pacients
```

```python
pacients['Last_name'] = pacients['Name'].str.split().str[-1]
pacients
```

```python
pacients['rate_upper'] = np.where(pacients['Upper_pressure'] > 120, 'high', 'normal')
pacients
```

```python
pacients['rate_lower'] = np.where(pacients['Lower_pressure'] > 80, 'high', 'normal')
pacients
```

```python
pacients['rate_weight'] = np.where(pacients['Weight'] > 100, 'overweight', 'normal')
pacients
```
