```python
import numpy as np
import matplotlib.pyplot as plt
import skimage 
astronaut = skimage.data.astronaut()
rocket = skimage.data.rocket()
rocket = rocket[0:400,0:600]
astronaut = astronaut[0:400,0:500]

merge = np.concatenate((astronaut,rocket), axis = 1)
flipped = np.fliplr(merge)
double = np.concatenate((merge,flipped),axis = 0)
#plt.gray()
plt.imshow(double)
```
