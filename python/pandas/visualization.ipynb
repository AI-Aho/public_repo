#### Wordcloud of category

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

text = " ".join(review for review in data.category.astype(str))
stopwords = set(" ")

# Create the WordCloud object
wordcloud = WordCloud(width=800, height=800, background_color='white').generate(text)

# Plot the WordCloud
plt.figure(figsize=(8, 8), facecolor=None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```

#### Price and category

```python
sns.set(rc={'figure.figsize':(18,8.27)})
sns.barplot(x = 'price',
            y = 'category',
            data = data,
            estimator = np.median,
            ci = 0)
 
# Show the plot
plt.show()
```

#### Gender and age

```python
sns.set(rc={'figure.figsize':(18,8.27)})
sns.barplot(x = 'gender',
            y = 'age',
            data = data,
            estimator = np.sum,
            ci = 0)
 
# Show the plot
plt.show()
```

#### Shopping mall and price

```python
sns.set(rc={'figure.figsize':(18,8.27)})
sns.barplot(x = 'shopping_mall',
            y = 'price',
            data = data,
            estimator = np.sum,
            ci = 0)
 
# Show the plot
plt.show()
```
